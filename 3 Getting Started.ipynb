```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,5)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
nx.draw_networkx(G)
```

```python
# set positions of the nodes using a dictionary

pos = {1:(0,0), 2:(-1,0.35), 3:(2,0.18), 4:(4,0.255), 5:(5,0.06)}

nx.draw_networkx(G, pos)
```

```python
options = {"font_size":32, "node_size":3000, "node_color":"white", 
           "edgecolors":"black", "linewidths":2, "width":5}
# linewidths is for symbol border, and width is for edge width

nx.draw_networkx(G, pos, **options)
# ** takes a dictionary and extends it into a series of arguments
```

```python
options = {"font_size":32, "node_size":3000, "node_color":"white", 
           "edgecolors":"black", "linewidths":2, "width":5}
# linewidths is for symbol border, and width is for edge width

nx.draw_networkx(G, pos, **options)
# ** takes a dictionary and extends it into a series of arguments

ax = plt.gca()
# gca is to get the current axes
ax.margins(0.2)
# set both x axis margin and y axis margins to 0.2 so that the display is not cut off
plt.axis("on")
plt.show()
```
