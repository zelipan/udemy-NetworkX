```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# The DiGraph class provides additional methods and properties specific to directed edges
```

```python
DG = nx.DiGraph()
DG.add_weighted_edges_from([(1,2,0.8), (3,1,0.4), (1,5,0.9), (1,4,0.2)])
nx.draw_networkx(DG)
```

```python
DG.out_degree(1)
# how many edges leaving 1
```

```python
DG.out_degree(1, weight='weight')
# total weight of edges leaving node 1
```

```python
DG.degree(1, weight='weight')
# total in degree and out degree weight
```

```python
list(DG.successors(1))
```

```python
list(DG.predecessors(1))
```

```python
H = nx.Graph(DG)
# create an undirected graph from a directed graph

nx.draw_networkx(H)
```
