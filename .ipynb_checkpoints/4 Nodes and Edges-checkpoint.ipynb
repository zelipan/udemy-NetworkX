```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Nodes

```python
G = nx.Graph()
G.add_node(1)

G.add_nodes_from([2,3])
# add nodes based on a list

G.add_nodes_from([ (5,{"color":"red"}), (4,{"color":"green"}) ])
# add nodes based on tuples in the format of (node, {node_attribute_ditionary})
# note that we cannot visualize the colors;
# the colors are only attributes of the nodes;
# we use node 5 to represent red;
# to display the color, we can use **options learned in the previous lesson

nx.draw_networkx(G)
```

##### Graph attributes

```python
G = nx.Graph(day="Friday")
G.graph
# set the attribute of the graph
```

```python
G.graph["day"] = "Tuesday"
G.graph
# modify attribute
```

##### Node attributes

```python
G.add_node(1, time="5pm")
# node 1 represents 5 pm

G.add_nodes_from([3], time="1pm")

G.nodes[1]["room"] = 812
# use node 1 to represent hotel room 812 as well

G.nodes.data()
```

##### Nodes from one graph can be incorporated into another

```python
T = nx.path_graph(10)
nx.draw_networkx(T)
```

```python
G.add_nodes_from(T)
nx.draw_networkx(G)
```

### Edges

```python
G = nx.Graph()
G.add_edge(1,2)

e = (2,3)
# we can also define an edge with a variable

G.add_edge(*e)
# unpack a tuple with 1 asterisk
# if we want to unpack a dictionary, we use 2 asterisks, as we have seen **options in the previous lesson

nx.draw_networkx(G)
```

```python
T = nx.path_graph(10)
G.add_edges_from(T.edges)
# add edges from another network T

nx.draw_networkx(G)
```

```python
G.clear()
# to empty a graph

G.nodes.data()
```

```python
G.add_edges_from([(1,2), (1,3)])

G.add_node(1)
G.add_edge(1,2)
# if we add new nodes/edges, NetworkX quietly ignores any that are already present

G.add_node("paper")
# we can also add a node with data type other than an integer

nx.draw_networkx(G)
```

```python
H = nx.Graph()

H.add_nodes_from("paper")
# if we pass a string in add_nodes_from, NetworkX will interpret paper as a list
# since the 1st letter p is already added, the 2nd letter p will not be ignored because p is already present

nx.draw_networkx(H)
```

##### Edge attributes

```python
G = nx.Graph()

G.add_edge(1,2,weight=5.3)
# adding attribute weight

G.add_edges_from([(3,4), (4,5)], color="green")
G.add_edges_from([(1,2,{"color":"blue"}), (2,3,{"weight":8})])

G.edges[2,3]["color"] = "red"
G.edges[3,4]["weight"] = 6.7
# setting the weight value of edge 3-4
G.edges[4,5]["weight"] = 11

G.edges.data()
```
