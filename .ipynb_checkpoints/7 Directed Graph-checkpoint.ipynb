```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
H = nx.DiGraph()
H.add_edge(0,3)
# we can use the same functions to add edges as we do for undirected graph
nx.draw_networkx(H)
```

```python
G = nx.DiGraph([(0,3), (1,3), (2,4), (3,6), (5,6)])

nx.draw_networkx(G)
```

```python
left_nodes = [0,1,2]
middle_nodes = [3,4]
right_nodes = [5,6]
# group nodes by column

pos = {n: (0,i) for i,n in enumerate(left_nodes)}
# the enumerate function adds a counter

pos.update({n: (1,i+0.5) for i,n in enumerate(middle_nodes)})

for i,n in enumerate(right_nodes):
    pos.update({n: (2,i+0.5)})

print(pos)
```

```python
nx.draw_networkx(G, pos)
```

```python
# One application of column groupping:
# We use column groupping to better visualise a flow graph
# A flow network is a directed graph where each edge has a capacity and each edge receives a flow
# The amount of flow on an edge cannot exceed the capacity of that edge
```
