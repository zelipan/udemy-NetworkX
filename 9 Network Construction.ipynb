```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Network Construction

```python
G = nx.Graph()

G.add_edge("a", "b", weight=0.6)
G.add_edge("a", "c", weight=0.2)
G.add_edge("c", "d", weight=0.1)
G.add_edge("c", "e", weight=0.7)
G.add_edge("c", "f", weight=0.9)
G.add_edge("a", "d", weight=0.3)
```

```python
G.nodes(data=True)
```

```python
G.edges.data()
```

```python
large_edges = [(u,v) for (u,v,d) in G.edges(data=True) if d["weight"] >= 0.5]

large_edges
```

```python
small_edges = [(u,v) for (u,v,d) in G.edges.data() if d["weight"] < 0.5]

small_edges
```

```python
pos = nx.spring_layout(G, seed=7)
# uses Fruchterman-Reingold (a force-directed layout algorithm which minimizes the energy of the system)
```

```python
nx.draw_networkx_nodes(G, pos, node_size=650)
# draw nodes

nx.draw_networkx_edges(G, pos, edgelist=large_edges, width=6)
nx.draw_networkx_edges(G, pos, edgelist=small_edges, width=6, alpha=0.5, edge_color="blue", style="dashed")
# draw edges; alpha controls opacity

nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif")
# label nodes

edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
# label edges with weights
```
